```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset=pd.read_csv("Iris.csv")
```

```python
dataset.head()
```

```python
dataset=dataset.set_index("Id")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.replace(list(dataset["Species"].unique()),[0,1,2],inplace=True)
```

```python
list(dataset["Species"].unique())
```

```python
dataset.head()
```

```python
X=np.array(dataset.iloc[:,:4])
y=np.array(dataset.iloc[:,[False,False,False,False,True]])
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8)
```

```python
from sklearn.tree import DecisionTreeClassifier
Dtree=DecisionTreeClassifier(criterion="gini",random_state=0)
tree_model=Dtree.fit(X_train,y_train)
y_preds=tree_model.predict(X_test)
```

```python
X_test
```

```python
len(X_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(Dtree, X_train , y_train )
```

```python
plt.plot(np.arange(0,30),y_preds)
plt.plot(np.arange(0,30),y_test)
```
